```python
import pandas as pd

# Baca dataset
df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")

# Tampilkan informasi awal
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["Global_Sales"], bins=50, kde=True)
plt.xlabel("Penjualan Global (Juta Unit)")
plt.ylabel("Jumlah Game")
plt.title("Distribusi Penjualan Global Video Game")
plt.show()
```

```python
top_publishers = df.groupby("Publisher")["Global_Sales"].sum().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_publishers.values, y=top_publishers.index, palette="viridis")
plt.xlabel("Total Penjualan Global (Juta Unit)")
plt.ylabel("Publisher")
plt.title("Top 10 Publisher dengan Penjualan Tertinggi")
plt.show()
```

Dari semua data yang sudah ditampilkan kita dapat menyimpulkan bahwa nitendo memiliki total penjumlahan unit yang lebih besar dari produser lain secara global.
